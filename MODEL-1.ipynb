```python
data_copy = data.drop(['area_mean','area_worst','perimeter_mean'], axis=1)
```

```python
data_copy.head()
```

```python
#We will make another copy of the dataset and will only include numerical variables
X = data_copy.drop(['id','diagnosis'],axis=1)
```

```python
for feature in X.columns:
    print(feature)
    print(X[feature].skew())
    print("---------------")
```

```python
X.isnull().sum().sum()
```

```python
for feature in X.columns:
    X[feature] = np.where(X[feature]<=0,0,np.round(np.sqrt(X[feature]),4))
```

```python
X.isnull().sum().sum()
```

```python
for feature in X.columns:
    print(feature)
    print(X[feature].skew())
    print("---------------")
```

```python
for feature in X.columns:
    sns.distplot(X[feature])
    plt.show()
```

We still have two variables which are having skewed variable and needs to be handled
1. concavity_se
2. area_se

The concavity_se is not having that much skewness now but we will look into this later if needed

```python
#We will check the distribution of the area_se
X['area_se'].describe()
```

```python
X['area_se'] = np.round(np.log(X['area_se']),4)
```

```python
#We will plot the distributionplot for the area and see distribution
sns.distplot(X['area_se'])
```

This looks much better then the previous distribution

```python
#We will now check the outliers in the dataset
for feature in X.columns:
    sns.boxplot(X[feature])
    plt.show()
```

```python
#Now we will use flooring and capping method to handle the outliers
for feature in X.columns:
    X[feature] = np.where(X[feature]<X[feature].quantile(0.10),X[feature].quantile(0.10),X[feature])
    X[feature] = np.where(X[feature]>X[feature].quantile(0.90),X[feature].quantile(0.90),X[feature])
```

```python
for feature in X.columns:
    sns.boxplot(X[feature])
    plt.show()
```

Now there are no outliers in the datset. 

```python
#Converting the target variable into numerical
data['diagnosis'] = np.where(data['diagnosis']=='M',1,0)
```

```python
#Coping the trget variable into the new dataframe
y = data['diagnosis']
```

```python
#We will now check the shape of the datsets
X.shape,y.shape
```

```python
#Now we can perform train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state =1)
```

```python
#WE will check the shape of these
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
#We will now import the rest of the libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
#We will create the object
knn1= KNeighborsClassifier()
```

```python
knn1.fit(X_train,y_train)
```

```python
pred = knn1.predict(X_test)
```

```python
print(classification_report(y_test,pred))
print("---------------------------------")
print(accuracy_score(y_test,pred))
print("---------------------------------")
print(confusion_matrix(y_test,pred))
print("---------------------------------")
print(roc_auc_score(y_test,pred))
```

## With Hyper Parameter

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_neighbour = list(range(1,30))
leaf_size = list(range(30,50))
weight = ['uniform']
param = {'n_neighbors':n_neighbour, 'leaf_size':leaf_size,'weights':weight}
```

```python
knn2= KNeighborsClassifier()
grid = RandomizedSearchCV(knn2,param_distributions=param,n_iter=5)
```

```python
new_model= grid.fit(X_train,y_train)
```

```python
new_model.best_estimator_
```

```python
pred = grid.predict(X_test)
```

```python
print(classification_report(y_test,pred))
print("---------------------------------")
print(accuracy_score(y_test,pred))
print("---------------------------------")
print(confusion_matrix(y_test,pred))
print("---------------------------------")
print(roc_auc_score(y_test,pred))
```
